```python
import pandas as pd
df=pd.read_csv('/content/crypto_data.csv')
```

##Направете претпроцесирање на податоците

```python
df.isnull().sum()
```

```python
data=df
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
```

```python
data.head()
```

```python
data_copy = data
```

```python
data = data.sort_values(by='Date', ascending=False)
for i in range(1, 4):
  data[f'Close BTC from {i} days ago'] = data['Close (BTC)'].shift(i)

data.dropna(subset=data.columns,axis=0,inplace=True)

data.head(10)
```

##Поделето го на тренирачко и тестирачко множество во сооднос 80:20

```python
data = data.reset_index()
data.drop('Date',axis=1,inplace=True)
```

```python
data
```

```python
X, Y = data.drop('Close (BTC)',axis=1), data['Close (BTC)']
```

```python
#first 80% for train, last 20% for test
X_train = X.iloc[int(len(data)*0.2):,:]
X_test = X.iloc[:int(len(data)*0.2),:]
Y_train = Y[int(len(data)*0.2):]
Y_test = Y[:int(len(data)*0.2)]
```

##Искористете соодветен Boosting модел за проблемот и тестирајте го моделот на тестирачкото множество

```python
from xgboost import XGBRegressor
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(Y_test,Y_pred))
```

```python
#poblisku do 1 podobro
from sklearn.metrics import r2_score
print(r2_score(Y_test,Y_pred))
```

##Искористете соодветен Decision Tree модел за проблемот и тестирајте го моделот на тестирачкото множество

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth=100)

# Train Decision Tree Regressor
regressor.fit(X_train, Y_train)

# #Predict the response for test dataset
Y_pred = regressor.predict(X_test)
```

```python
mean_squared_error(Y_test, Y_pred)
```

```python
print(r2_score(Y_test,Y_pred))
```
